### 5) converting strings into list

```python
movies.iloc[0].genres
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
# applying this above concept in the below function for proper extraction
```

```python
def convert(obj):
    list = []
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list    
```

### 6) extracting keywords from genres column

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

### 7) extracting keywords from keywords column

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

### 8) extracting keywords from cast column ( top 3 actors only)

```python
movies.iloc[0].cast
```

```python
def convert_cast(obj):
    list = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            list.append(i['name'])
            counter+=1
        else:
            break
    return list 
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

### 9) extracting keywords from crew column ( director only)

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    list = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            list.append(i['name'])
    return list
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

### 10) converting overview column into list

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

### 11) removing space between the words

```python
movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

### 12) creating tag column

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

### 4) function for recomending top 5 movies

##### index of a movie

```python
new_df[new_df['title'] == 'Batman Begins'].index[0]
```

##### distances at 119 index in similarity matrix

```python
similarity[119]
```

##### sorting in reverse order without losing its index position

```python
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[119])), reverse=True,key=lambda x:x[1])
```

```python
def recomend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recomend('Avatar')
```
